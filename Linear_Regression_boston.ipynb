# Boston House Price Prediction

### Collecting data from Sklearn dataset

```python
from sklearn.datasets import load_boston  
```

```python
data=load_boston()
```

```python
x,y=load_boston(return_X_y=True)
```

### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Data preprocessing

```python
x=pd.DataFrame(x,columns=data.feature_names)
```

```python
y=pd.DataFrame(y,columns=['Price'])
```

```python
x
```

```python
y
```

```python
df=x[data.feature_names]
```

```python
df['Price']=y['Price']
```

###  Exploratory Data Analysis

```python
df.info()  #info about dataset features
```

```python
x_corr=x.corr()
```

```python
x_corr
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(x_corr,annot=True)
plt.show()
```

```python
columns=[]
for i in range(x_corr.shape[0]):
    for j in range(i):
        if abs(x_corr.iloc[i,j])>0.7:
             columns.append(x_corr.columns[i])
corr_features=set(columns)                
```

```python
corr_features
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
x_train.drop(corr_features,axis=1)
```

```python
x_test.drop(corr_features,axis=1)
```

### Linear Regression

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
model_error=mean_squared_error(y_test,model.predict(x_test))
RMSE=np.sqrt(model_error)
R2_score=r2_score(y_test,model.predict(x_test))
print('MSE error is {}'.format(model_error))
print('RMSE error is {}'.format(RMSE))
print('r2_score error is {}'.format(R2_score))
```

#### MSE should be nearer to zeo for best model performance

###  Ridge Regression

```python
from sklearn.linear_model import Lasso,Ridge
ridge=Ridge(alpha=0.0001)
ridge_model=ridge.fit(x_train,y_train)
Ridge_pred=ridge.predict(x_test)
R2=r2_score(y_test,Ridge_pred)
print('R2 score is {}'.format(R2))
```

###  Lasso Regression

```python
Las=Lasso(alpha=0.0001)
Las=Las.fit(x_train,y_train)
pred_las=Las.predict(x_test)
R2=r2_score(y_test,pred_las)
print('R2 score is {}'.format(R2))
```

###  Normal linear regression and regularized model have same r2_score so why model is good with accuracy have 78.36%
